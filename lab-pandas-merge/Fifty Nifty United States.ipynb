```python
import pandas as pd
```

```python
state_a=pd.read_csv('states-a.csv')
```

```python
# output from states-a.csv
```

```python
state_b=pd.read_csv('states-b.csv')
```

```python
# output from states-b.csv
```

```python
# after you combine the two, your dataframe should have this shape
```

```python
states=pd.concat([state_a,state_b],axis=0)
states.shape
```

```python
ab=pd.read_csv('state-abbreviations.csv')
```

```python
states=pd.merge(states,ab,on='State')
states.rename(columns={'Abbreviation_x':'Abbreviation'},inplace=True)
states.drop("Abbreviation_y",axis=1,inplace=True)
states.head()
```

```python
# output after combining state-abbreviations.csv
```

```python
cp=pd.read_csv('state-capitals.csv')
states=pd.merge(states,cp,on='State')
states.rename(columns={'Cpital_x':'Capital'},inplace=True)
states.drop("Capital_y",axis=1,inplace=True)
states.head()
```

```python
# output after combining state-capitals.csv
```

```python
states.set_index('State',inplace=True)
```

```python
states.head()
```

```python
# output after setting the index to be the State column
```

```python
fc=pd.read_csv('state-facts.csv')
fc.set_index('State',inplace=True)
```

```python
fc.head()
```
